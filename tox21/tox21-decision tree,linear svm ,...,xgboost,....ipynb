#### initial library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option("display.max_columns",None)
```

```python
features=pd.read_csv('tox21_global_cdf_rdkit.csv')
features=features.drop("('RDKit2DNormalized_calculated', <class 'bool'>)",axis=1)
x=features
```

#### imputing for features

```python
x.head()
```

```python
x.head()
```
